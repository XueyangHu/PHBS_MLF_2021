```python
from scipy.io import arff
import pandas as pd
import numpy as np
```

```python
data = arff.loadarff('4year.arff')
df = pd.DataFrame(data[0])
```

```python
df['bankruptcy'] = (df['class']==b'1')
```

```python
df.head()
```

```python
df.describe()
```

```python
# Create a new dataframe with only 4 feataures and Bankruptcy
df2 = df[['Attr1', 'Attr2', 'Attr7', 'Attr10', 'class', 'bankruptcy']]
```

```python
# Rename columns
df2.columns = ['X1', 'X2', 'X7', 'X10', 'class', 'bankruptcy']
df2
```

```python
df2.isna().sum()
```

```python
# Fill-in the missing values (nan) with the column means
df2.fillna(value=df2.mean(axis=0), inplace=True)
df2
```

```python
# Find the mean and std of the 4 features among all, bankrupt and still-operating companies
mean = df2.groupby(['bankruptcy']).mean().append(pd.Series(df2.mean()), ignore_index=True).drop(axis=1, columns=['class', 'bankruptcy'])
std = df2.groupby(['bankruptcy']).std().append(pd.Series(df2.std()), ignore_index=True).drop(axis=1, columns=['bankruptcy'])
```

```python
mean.index = ['still-operating', 'bankrupt', 'all']
print("Mean:")
print(mean)
```

```python
std.index = ['still-operating', 'bankrupt', 'all']
print("Std:")
print(std)
```

```python
# Bankrupt companies
df2.loc[df2['bankruptcy']==True].index
```

```python
# Still-operating companies
df2.loc[df2['bankruptcy']==False].index
```

```python
# How many companies satisfy the condition, X1 < mean(X1) - stdev(X1) AND X10 < mean(X10) - std(X10)
x1_x10 = df2.loc[(df2['X1'] < df2['X1'].mean() - df2['X1'].std()) & (df2['X10'] < df2['X10'].mean() - df2['X10'].std())]
x1_x10
```

```python
len(x1_x10)
```

```python
# The ratio of the bankrupted companies among the sub-groups above
ratio = sum(x1_x10['bankruptcy']==True) / len(x1_x10) 
ratio
```
